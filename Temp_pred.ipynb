```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\PC1\Desktop\DS\DataSets\temperatures.csv")
```

```python
df
```

```python
#input variable for graph
x = df['YEAR']

#output variable for graph
y = df['ANNUAL']
```

```python
plt.figure(figsize=(16,9))
plt.title('Avg. Temperature Plot of INDIA')
plt.xlabel('Year')
plt.ylabel('Avg. Annual')
plt.errorbar(x,y)
plt.scatter(x,y)
```

```python
x = x.values
```

```python
x = x.reshape(117,1) #Requirement of python library  
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x,y) #Done y=mx+c
```

```python
regressor.coef_  #m 
```

```python
regressor.intercept_ #c
```

```python
regressor.predict([[2030]])
```

```python
predicted = regressor.predict(x)
```

```python
predicted 
```

```python
y
```

```python
#Mean Absolute Error
np.mean(abs(y - predicted))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y, predicted)
```

```python
#Mean Squared Error
np.mean((y - predicted)**2)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, predicted)
```

```python
from sklearn.metrics import r2_score
r2_score(y, predicted)
```

```python
regressor.score(x,y)
```

```python
plt.figure(figsize=(16,9))
plt.title('Avg. Temperature Plot of INDIA')
plt.xlabel('Year')
plt.ylabel('Avg. Annual')
plt.scatter(x,y, label = 'actual', color = 'r', marker = '+')
plt.errorbar(x, predicted, label = 'predicted', color = 'c')
plt.legend() #To display which color denote what
```

```python
plt.figure(figsize=(16,9))
plt.title('Avg. Temperature Plot of INDIA')
plt.xlabel('Year')
plt.ylabel('Avg. Annual')
plt.errorbar(x,y, label = 'actual', color = 'y')
plt.plot(x, predicted, label = 'predicted', color = 'c')
plt.legend() #To display which color denote what
```
